# Funciones algebraicas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

## 1. Funci√≥n lineal 

```python
n = 100

x1 = 0
y1 = 1
```

```python
x = np.linspace(-10, 10, num=n)
```

```python
def find_equation_of_line(m, x, b):
    return m * x + b

def find_slope(x2, y2):
    return (y2 - y1) / (x2 - x1)
```

Ejemplos:
$$ y = x + 1 $$
$$ y = 0.33X + 1 $$
$$ y = -0.17X + 1 $$

```python
m = round(find_slope(1, 2), 2)
b = y1 - m * x1
y = find_equation_of_line(m, x, b)


plt.figure(figsize=(15, 10))

sns.lineplot(data= y, color='skyblue', label=f"f(x) = {m}x + {b}")

m = round(find_slope(3, 2), 2)
b = y1 - m * x1
y = find_equation_of_line(m, x, b)

sns.lineplot(data= y, color='salmon', label=f"f(x) = {m}x + {b}")


m = round(find_slope(3, 0.5), 2)
b = y1 - m * x1
y = find_equation_of_line(m, x, b)

sns.lineplot(data= y, color='lightgreen', label=f"f(x) = {m}x + {b}")
```

## 2. Funci√≥n polin√≥mica 

Ejemplo:
$$ 2X^7 - X^4 + 3X^2 + 4 $$

```python
def examplePolynomial(x):
    return 2 * x ** 7 - 4 * x ** 3 + 3 * x **2 + 4
```

```python
y = examplePolynomial(x)

plt.figure(figsize=(15, 10))
sns.lineplot(data= y, color='lightgreen')
```

# Funciones trascendentes

## 1. Funciones trigonom√©tricas

Ejemplo:

$$ sin(x) $$
$$ cos(x) $$

```python
def cos(x):
    return np.cos(x)
    
def sin(x):
    return np.sin(x)
```

```python
y = cos(x)

plt.figure(figsize=(15, 10))

sns.lineplot(data= y, color='salmon', label=f"cos(x)")

y = sin(x)
sns.lineplot(data= y, color='lightgreen', label=f"sin(x)")
```

# Ingresos y gastos durante un año para un Local Comercial
Estás llevando un registro de tus gastos mensuales y quieres visualizar cómo han variado a lo largo de los meses. Con el objetivo de llevar un registro detallado de tus finanzas y tomar decisiones más informadas, decides mantener un registro mensual de tus ingresos y gastos.

Durante todo el año, meticulosamente anotas cada transacción. En enero, los ingresos son modestos ya que el centro comercial aún se está recuperando de la temporada de vacaciones, pero tus gastos son mínimos mientras ajustas tu inventario para el nuevo año.

Conforme avanza el año, tus ingresos comienzan a aumentar gradualmente a medida que el tráfico de clientes en el centro comercial se estabiliza. Sin embargo, los gastos también aumentan a medida que inviertes en publicidad, mejoras en el local y nuevas existencias para mantener la tienda fresca y atractiva para los clientes.

Llegado diciembre, te sientes satisfecho con el progreso de tu negocio, pero también te das cuenta de la importancia de analizar tus finanzas para planificar mejor el próximo año.

Ahora, con una visión más clara de tus finanzas, te dispones a calcular el total de ingresos y gastos del año, así como los valores máximo, mínimo y promedio. Esta información te ayudará a evaluar el desempeño de tu negocio y a establecer metas financieras más realistas para el próximo año.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Meses
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Gastos mensuales
gastos = [50, 150, 200, 180, 220, 190, 210, 220, 230, 240, 310, 480]

# Ingresos mensuales
ingresos = [100, 250, 220, 210, 240, 230, 250, 260, 270, 280, 320, 605]

# Crear el gráfico
plt.plot(meses, gastos, marker='o', linestyle='-', label='Gastos', color='red')
plt.plot(meses, ingresos, marker='o', linestyle='-', label='Ingresos', color='green')


# Etiquetas de los ejes y título
plt.xlabel('Mes')
plt.ylabel('Dinero (COP)')
plt.title('Comparación de ingresos y gastos mensuales - Inventas App')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.gcf().set_size_inches(5, 5)
plt.show()
```

### Ahora necesitamos saber cuanto es el total de gastos e ingresos, asi como el promedio de estos

```python
promedio_gastos = np.mean(gastos)
max_gastos = np.max(gastos)
min_gastos = np.min(gastos)

# Etiquetas para las barras
labels = ['Promedio', 'Máximo', 'Mínimo']

# Valores para las barras
values = [promedio_gastos, max_gastos, min_gastos]

# Crear el gráfico de barras
plt.bar(labels, values, color=['blue', 'green', 'red'])

plt.xlabel('Dinero (COP)')
plt.ylabel('Valor de gastos')
plt.title('Comparación de promedio, máximo y mínimo de gastos')

# Mostrar el gráfico
plt.gcf().set_size_inches(5, 5)
plt.show()
```

```python
promedio_ingresos = np.mean(ingresos)
max_ingresos = np.max(ingresos)
min_ingresos = np.min(ingresos)

# Etiquetas para las barras
labels = ['Promedio', 'Máximo', 'Mínimo']

# Valores para las barras
values = [promedio_ingresos, max_ingresos, min_ingresos]

# Crear el gráfico de barras
plt.bar(labels, values, color=['blue', 'green', 'red'])

plt.xlabel('Dinero (COP)')
plt.ylabel('Valor de ingresos')
plt.title('Comparación de promedio, máximo y mínimo de gastos')

# Mostrar el gráfico
plt.gcf().set_size_inches(5, 5)
plt.show()
```

```python
# Crear el histograma de gastos
plt.hist(gastos, bins=10, color='red', alpha=0.5, label='Gastos')

# Crear el histograma de ingresos
plt.hist(ingresos, bins=10, color='blue', alpha=0.5, label='Ingresos')

# Añadir etiquetas y título
plt.xlabel('Monto')
plt.ylabel('Frecuencia')
plt.title('Histograma de gastos e ingresos')
plt.legend()

# Mostrar el histograma
plt.gcf().set_size_inches(5, 5)
plt.show()
```

```python
diff_ingresos_gastos = [ingresos[i] - gastos[i] for i in range(len(ingresos))]

promedio_anual_gastos  = np.mean(diff_ingresos_gastos)
min_anual_gastos  = np.min(diff_ingresos_gastos)
max_anual_gastos  = np.max(diff_ingresos_gastos)

# Crear gráfico de barras para comparar gastos de cada mes con el promedio anual
plt.bar(meses, diff_ingresos_gastos, color='green', alpha = 0.5, label='Ganancias')

plt.axhline(y=promedio_anual_gastos, color='blue', linestyle='--', label='Promedio ganancias')
plt.axhline(y=max_anual_gastos, color='yellow', linestyle='--', label='Máximo gastos')
plt.axhline(y=min_anual_gastos, color='purple', linestyle='--', label='Minimo gastos')
plt.xlabel('Mes')
plt.ylabel('Dinero (COP) K')
plt.title('Ganancias totales por mes')
plt.legend()
plt.xticks(rotation=45)
plt.gcf().set_size_inches(5, 5)
plt.show()
```

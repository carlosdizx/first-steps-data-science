# Iris dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Iris.csv')
```

```python
sns.pairplot(df, hue='Species')
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.histplot(df['SepalLengthCm'], kde=True)
plt.title('Distribucin de la longitud del Spalo')
plt.xlabel('Longitud del Spalo (cm)')
plt.ylabel('Frecuencia')

plt.subplot(2, 2, 2)
sns.histplot(df['SepalWidthCm'], kde=True)
plt.title('Distribucin del Ancho del Spalo')
plt.xlabel('Ancho del Spalo (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.histplot(df['PetalLengthCm'], kde=True)
plt.title('Distribucin de la longitud del Ptalo')
plt.xlabel('Longitud del Spalo (cm)')
plt.ylabel('Frecuencia')

plt.subplot(2, 2, 2)
sns.histplot(df['PetalWidthCm'], kde=True)
plt.title('Distribucin del Ancho del Ptalo')
plt.xlabel('Ancho del Spalo (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

# No nos interesa hacer anlisis sobre las dimensiones del ptalo!!

### Regresin lineal

```python
X = df['PetalLengthCm']
Y = df['PetalWidthCm']
N = len(X)
sum_x = X.sum()
sum_y = Y.sum()
sum_xy = (X * Y).sum()
sum_x_squared = (X ** 2).sum()

m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
b = (sum_y - m * sum_x) / N

print(f'Pendiente (m) calculada manualmente: {m}')
print(f'Interseccin (b) calculada manualmente: {b}')
```

### Regresin lineal con librera sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['PetalLengthCm']]
Y = df[['PetalWidthCm']]

regresor = LinearRegression()
regresor.fit(X, Y)

sklearn_m = regresor.coef_[0][0]
sklearn_b = regresor.intercept_[0]

print(f'Pendiente (m) calculada con scikit-learn: {sklearn_m}')
print(f'Interseccin (b) calculada con scikit-learn: {sklearn_b}')
```

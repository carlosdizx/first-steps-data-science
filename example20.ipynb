# Iris dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Iris.csv')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['SepalLengthCm'], kde=True, ax=axes[0])
axes[0].set_title('Distribución de la Longitud del Sépalo')
axes[0].set_xlabel('Longitud del Sépalo (cm)')
axes[0].set_ylabel('Frecuencia')

sns.histplot(df['SepalWidthCm'], kde=True, ax=axes[1])
axes[1].set_title('Distribución del Ancho del Sépalo')
axes[1].set_xlabel('Ancho del Sépalo (cm)')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['PetalLengthCm'], kde=True, ax=axes[0])
axes[0].set_title('Distribución de la Longitud del pétalo')
axes[0].set_xlabel('Longitud del pétalo (cm)')
axes[0].set_ylabel('Frecuencia')

sns.histplot(df['PetalWidthCm'], kde=True, ax=axes[1])
axes[1].set_title('Distribución del Ancho del pétalo')
axes[1].set_xlabel('Ancho del petalo (cm)')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

# No nos interesa hacer análisis sobre las dimensiones del pétalo!!

### Regresión lineal

```python
X = df['PetalLengthCm']
Y = df['PetalWidthCm']
N = len(X)
sum_x = X.sum()
sum_y = Y.sum()
sum_xy = (X * Y).sum()
sum_x_squared = (X ** 2).sum()

m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
b= (sum_y - m * sum_x) / N

print(f'Pendiente (m) calculada con scikit-learn: {m}')
print(f'Intersección (b) calculada con scikit-learn: {b}')
```

### Regresión lineal con librería sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['PetalLengthCm']]
Y = df[['PetalWidthCm']]

regresor = LinearRegression()
regresor.fit(X, Y)

sklearn_m = regresor.coef_[0][0]
sklearn_b = regresor.intercept_[0]

print(f'Pendiente (m) calculada con scikit-learn: {sklearn_m}')
print(f'Intersección (b) calculada con scikit-learn: {sklearn_b}')
```

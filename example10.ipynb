# Análisis del rendimiento de un equipo de fútbol

Eres un apasionado seguidor de un equipo de fútbol local y te encuentras sumergido en el análisis del rendimiento del equipo durante la última temporada. Decides llevar un registro detallado de los goles marcados y recibidos en cada partido, así como de los resultados obtenidos.

A lo largo de la temporada de 20 emocionantes partidos, registras meticulosamente cada detalle. Desde las victorias eufóricas hasta los empates frustrantes y las derrotas desalentadoras, capturas la esencia de cada encuentro en tus datos.

Con la ayuda de Matplotlib, decides visualizar tus datos para obtener una comprensión más clara del desempeño del equipo. Primero, creas un gráfico de barras que muestra los goles marcados y recibidos en cada partido. Observas cómo los altibajos en la puntuación reflejan los momentos de triunfo y adversidad a lo largo de la temporada.

Luego, recurres a un gráfico de pastel para examinar la distribución de los resultados de los partidos. Observas con fascinación cómo los porcentajes de victorias, empates y derrotas se entrelazan para formar la narrativa de la temporada.

Al final del análisis, te encuentras con una comprensión más profunda del rendimiento del equipo. Aunque los altibajos son inevitables en el mundo del fútbol, te sientes inspirado por la pasión y la dedicación del equipo para enfrentar cada desafío con determinación.

Con una nueva perspectiva en mente, te preparas para apoyar al equipo en la próxima temporada con aún más entusiasmo y optimismo.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos
partidos = list(range(1, 21))
goles_marcados = [2, 3, 1, 2, 4, 2, 3, 3, 2, 1, 2, 3, 4, 2, 1, 3, 2, 2, 3, 1]
goles_recibidos = [1, 0, 2, 1, 2, 1, 1, 2, 2, 2, 1, 0, 1, 2, 3, 1, 2, 1, 0, 2]

def calculoResultadoPartidos():
    resultadoPartidos = []
    for i in range(len(goles_marcados)):
        if goles_marcados[i] > goles_recibidos[i]:
            resultadoPartidos.append("Victoria")
        elif goles_marcados[i] == goles_recibidos[i]:
            resultadoPartidos.append("Empate")
        else:
            resultadoPartidos.append("Derrota")
    return resultadoPartidos

resultados = calculoResultadoPartidos()
    
```

```python
# Gráfico de barras para los goles marcados y recibidos en cada partido
plt.bar(partidos, goles_marcados, color='green', alpha=0.5, label='Goles marcados')
plt.bar(partidos, goles_recibidos, color='red', alpha=0.5, label='Goles recibidos')
plt.xlabel('Partido')
plt.ylabel('Goles')
plt.title('Goles marcados y recibidos por partido')
plt.legend()
plt.gcf().set_size_inches(5, 5)
plt.show()
```

```python
# Gráfico de pastel para los resultados de los partidos
victorias = resultados.count('Victoria')
empates = resultados.count('Empate')
derrotas = resultados.count('Derrota')
sizes = [victorias, empates, derrotas]
labels = ['Victorias', 'Empates', 'Derrotas']
colors = ['green', 'yellow', 'red']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Resultados de los partidos')
plt.gcf().set_size_inches(5, 5)
plt.show()
```

```python
prom_goles_marcados = np.mean(goles_marcados)
max_goles_marcados = np.max(goles_marcados)
min_goles_marcados = np.min(goles_marcados)

values = [prom_goles_marcados, max_goles_marcados, min_goles_marcados]
labels = ['Promedio', 'Máximo', 'Minimo']

plt.bar(labels, values, color=['yellow', 'purple', 'red'])

plt.ylabel('Goles')

plt.gcf().set_size_inches(5, 5)
plt.show()
```

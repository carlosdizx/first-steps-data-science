```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    # 'Mes': ,
    'Mes': np.random.choice(['Ene', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dic'], 100),
    'Gastos': [math.floor(x*1000000 + 30000000) for x in np.random.rand(100).data.tolist()],
    'Ingresos': [math.floor(x*2000000 + 40000000) for x in np.random.rand(100).data.tolist()],
})

data.head(-1)
```

```python
sns.barplot(x='Mes', y='Gastos', data=data)
plt.title("Gastos x mes")
plt.show()

sns.violinplot(x='Mes', y='Gastos', data=data)
plt.title("Gastos x mes")
plt.show()
```

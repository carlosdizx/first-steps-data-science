```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value': np.random.rand(100)
})

data.head(-1)
```

```python
sns.barplot(x='Category', y='Value', data=data )
plt.title("Graphic of Bars")
plt.show()

sns.barplot(y='Category', x='Value', data=data, orient='h')
plt.title("Graphic of Bars")
plt.show()
```

```python
data_scatterplot= pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

data_scatterplot.head(-1)
```

```python
sns.scatterplot(x='X', y='Y',hue='Category' , data=data_scatterplot)
plt.grid(False)
plt.title("Graphic Scatterplot")
plt.show()
```

```python
sns.histplot(data['Value'], bins=20, kde=True)
```

```python
sns.violinplot(x='Category', y='Value', data=data)
```

```python
data_matrix = np.random.rand(10, 10)
sns.heatmap(data_matrix, annot=True)
```

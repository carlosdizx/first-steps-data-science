```python
import pandas as pd
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Cargar el archivo CSV en un DataFrame de pandas
us_arrests = pd.read_csv("data/USArrests.csv")
us_arrests.head()
```

```python
scaler = StandardScaler()
arrests_scaled = scaler.fit_transform(us_arrests)

# Calcular la matriz de distancia euclidiana
distance_matrix = pdist(arrests_scaled, metric="euclidean")

hc = linkage(distance_matrix, method="ward")

# Crear el dendrograma
dendrogram(hc, labels=us_arrests.index, color_threshold=5)

# Línea horizontal para el corte
plt.axhline(y=5, color="red", linestyle="--")
plt.ylabel("Height")
plt.title("Clustering Jerárquico con Método Ward")
plt.show()
```

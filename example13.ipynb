```python
import plotly.express as px
import pandas as pd
```

```python
data_map = pd.DataFrame({
    'City': ['Madrid', 'Lima', 'Jhansi'],
    'Latitude': [40.4168, -12.04637310, 25.436298],
    'Longitude': [-3.7038, -77.04275400, 78.567352],
    'Value': [1,2,3]
})

fig = px.scatter_geo(data_map, 
                     lat='Latitude', lon='Longitude', hover_name='City', projection='natural earth', size='Value')

fig.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm


X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

plt.gcf().set_size_inches(5, 5)
plt.show()
```

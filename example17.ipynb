# Composici√≥n de funciones

$$ f(x)$$
$$g(x)$$
$$y = f(g(x))$$

Ejemplo

 $$ f(x) = x^2 $$
 $$ g(x) = sin(x) $$
 $$ y = f(g(x)) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 1000
x = np.linspace(-10,10, num=N)
```

```python
def f(x): return x ** 2

def g(x): return np.sin(x)
```

```python
plt.figure(figsize=(15, 10))

y= f(g(x))

sns.lineplot(data= y, color='skyblue', label="f(g(x))")
```

```python
plt.figure(figsize=(15, 10))

y= g(f(x))

sns.lineplot(data= y, color='skyblue', label="g(f(x))")
```

```python
plt.figure(figsize=(15, 10))

y= f(g(x))
sns.lineplot(data= y, color='skyblue', label="f(g(x))")

y= g(f(x))
sns.lineplot(data= y, color='lightgreen', label="g(f(x))")
```

```python
plt.figure(figsize=(15, 10))

y= g(x)
sns.lineplot(data= y, color='lightgreen', label="g(x)")


y= f(g(x))
sns.lineplot(data= y, color='skyblue', label="f(g(x))")
```

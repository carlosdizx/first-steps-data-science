```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/nhanes_2015_2016.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
sns.histplot(df, x="BPXSY1", kde=True)

plt.subplot(2, 2, 2)
sns.histplot(df, x="RIDAGEYR", kde=True)
plt.show()
```

```python
sns.relplot(y="BPXSY1", x="RIDAGEYR", alpha=0.5, data=df, height=5, aspect=3, hue="RIAGENDR")
plt.show()
```

```python
sns.relplot(y="BMXBMI", x="RIDAGEYR", alpha=0.5, data=df, height=5, aspect=3, hue="RIAGENDR")
plt.show()
```

```python
sns.relplot(y="BPXSY1", x="RIDAGEYR", alpha=0.5, data=df, height=5, aspect=3, hue="RIAGENDR")
plt.show()
```

```python
sns.relplot(y="BPXSY1", x="RIDAGEYR", alpha=0.5, data=df[df["RIAGENDR"] == 1], height=5, aspect=3)
plt.show()
```

```python
sns.relplot(y="BPXSY1", x="RIDAGEYR", alpha=0.5, data=df[df["RIAGENDR"] == 2], height=5, aspect=3)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(y="BPXSY1", x="RIDAGEYR", size="BPXSY2", sizes=(40, 400), data=df, hue="RIAGENDR", alpha=0.5)
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x="RIDAGEYR", y="BPXSY1", hue="RIAGENDR", data=df)
plt.title("Blood Pressure (Systolic) by Age and Gender")
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure (BPXSY1)")
plt.legend(title="Gender")
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x="RIDAGEYR", y="BPXSY2", hue="RIAGENDR", data=df)
plt.title("Blood Pressure (Systolic) by Age and Gender")
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure (BPXSY1)")
plt.legend(title="Gender")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x="BMXBMI", hue="RIAGENDR", kde=True, element="step", stat="density", common_norm=False)
plt.title("BMI Distribution by Gender")
plt.xlabel("Body Mass Index (BMI)")
plt.ylabel("Density")
plt.legend(title="Gender", labels=["Men", "Women"])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lmplot(x="BMXWT", y="BMXBMI", hue="RIAGENDR", data=df, aspect=1.5, markers=["o", "s"], palette="Set1")
plt.title("Relationship between Weight and BMI by Gender")
plt.xlabel("Weight (kg)")
plt.ylabel("Body Mass Index (BMI)")
plt.show()
```

```python
plt.figure(figsize=(14, 8))
age_marital = df.groupby(['RIDAGEYR', 'DMDMARTL']).size().unstack().fillna(0)
age_marital.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
plt.title("Marital Status Distribution by Age Group")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Marital Status")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x="RIDRETH1", y="BPXSY1", data=df, inner="quartile", hue="RIAGENDR")
plt.title("Distribution of Systolic Blood Pressure by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Systolic Blood Pressure (BPXSY1)")
plt.show()
```

```python
from scipy.spatial import distance
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Coordenadas de los puntos x e y
x = (1, 6, 2, 5, 3)
y = (3, 5, 2, 6, 6)

# Calcular las distancias
manhattan_dist = distance.cityblock(x, y)  # Distancia Manhattan
euclidean_dist = distance.euclidean(x, y)  # Distancia Euclidiana
chebyshev_dist = distance.chebyshev(x, y)  # Distancia Chebyshev

manhattan_dist, euclidean_dist, chebyshev_dist
```

```python
# Coordenadas de los puntos A, B, C y D
A = (2, 34)
B = (8, 40)
C = (0, 58)
D = (10, 10)
```

Ósea que X = (2, 8, 0, 10) e Y = (34, 40, 58, 10)

```python
# Calcular la distancia Euclidiana entre C y D
euclidean_distance_CD = distance.euclidean(C, D)
euclidean_distance_CD
```

```python
# Calcular la distancia Euclidiana entre A y B
chebyshev_distance_AB = distance.chebyshev(A, B)
chebyshev_distance_AB
```

```python
# Calcular el centroide
centroid_x = (A[0] + B[0] + C[0] + D[0]) / 4
centroid_y = (A[1] + B[1] + C[1] + D[1]) / 4

# Suma de las coordenadas x e y del centroide
centroid_sum = centroid_x + centroid_y
centroid = (centroid_x, centroid_y)
centroid_x, centroid_y, centroid_sum
```

```python
# Definir los puntos
points = {"A": A, "B": B, "C": C, "D": D}

# Crear el gráfico
plt.figure(figsize=(8, 8))
sns.scatterplot(x=[p[0] for p in points.values()], y=[p[1] for p in points.values()], s=100, color="blue", label="Puntos A, B, C, D")
sns.scatterplot(x=[centroid[0]], y=[centroid[1]], s=150, color="red", label="Centroide")


# Anotar los puntos y el centroide
for label, point in points.items():
    plt.text(point[0] + 0.3, point[1], f"{label} {point}", fontsize=10, color="blue")

plt.text(centroid[0] + 0.3, centroid[1], f"Centroide {centroid}", fontsize=10, color="green")

# Etiquetas y límites
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.title("Puntos y Centroide")
plt.legend()
plt.grid(True)
plt.show()
```
